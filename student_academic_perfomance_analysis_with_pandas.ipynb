### Challenge 2: Loading the Dataset

```python
# Load the Dataset from Github
url = "https://raw.githubusercontent.com/ritaafrica/data/refs/heads/main/student_scores.csv"
df = pd.read_csv(url)
```

### Challenge 3: Adding a New Column for Total Scores

```python
# a New column for Total Score
df["Total_Score"] = df["Math_Score"] + df["Reading_Score"] + df["Writing_Score"]
```

```python
df.head()
```

### Challenge 4: Adding a New Column to Calulate the Percentage Score

```python
# Calculating the Percentage Score
df["Percentage_Score"] = (df["Total_Score"] / 300) * 100
```

```python
df["Percentage_Score"] = df["Percentage_Score"].round(2)
```
